   # 2D versus 3D LES simulation of ASTEX stratocumulus

```python
# setting up everything
import numpy as np
import matplotlib.pyplot as plt
import xarray as xr

from matplotlib import cm

%matplotlib inline

path1="G:/My Drive/MyGit/SAM/data"

file="ASTEX_cold2.6_3D_120x120x512_35m_DRZ_RRTM_N100_3D.nc"

f1  = xr.open_dataset(path1+"/"+file)
```

```python
x = f1['time']
y = f1['LHF']

plt.plot(x,y)
```

```python
#plt.axis([165, 166, 0, 200])
plt.title("LHF time series")
plt.xlabel("days")
plt.ylabel("LHF")
plt.plot(x,y, color="darkorange")
```

```python
x = f1['time']
y = f1['z']
var = f1['QC']
var = var.T

plt.contourf(x,y,var, cmap = cm.Greys)
ax = plt.gca()
ax.set_ylim([0., 2000])

plt.savefig("Fig1_try.png")
```
